```python

import pandas as pd
import pylab as pl
import numpy as np
import scipy.optimize as opt
import statsmodels.api as sm
from sklearn import preprocessing
'exec(% matplotlib inline)'
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import seaborn as sn
```

```python
df=pd.read_csv('/heart_failure_clinical_records_dataset.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.dropna(axis = 0, inplace = True)
print(df.head(), df.shape)
```

```python
laste = df['DEATH_EVENT'].plot()
plt.show(laste)
```

```python
df.head()
```

```python
df.head()
```

```python
x = np.asarray(df[['age','anaemia','creatinine_phosphokinase','diabetes','high_blood_pressure','platelets','sex','smoking','time']])
y = np.asarray(df['DEATH_EVENT'])
 
# normalization of the dataset
x = preprocessing.StandardScaler().fit(x).transform(x)
 
# Train-and-Test -Split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 4)
```

```python
print ('Train set:', x_train.shape,  y_train.shape)
print ('Test set:', x_test.shape,  y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
 
logreg = LogisticRegression()
logreg.fit(x_train, y_train)
y_pred = logreg.predict(x_test)
 
# Evaluation and accuracy
from sklearn.metrics import accuracy_score
 
print('')
print('Accuracy of the model in accuracy score is = ',accuracy_score(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
 
rf = RandomForestClassifier()
rf.fit(x_train, y_train)
 
score = rf.score(x_test,y_test)*100
print('Accuracy of the model is = ', score)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
 
cm = confusion_matrix(y_test, y_pred)
conf_matrix = pd.DataFrame(data = cm,
                           columns = ['Predicted:0', 'Predicted:1'],
                           index =['Actual:0', 'Actual:1'])
 
plt.figure(figsize = (8, 5))
sn.heatmap(conf_matrix, annot = True, fmt = 'd', cmap = "Greens")
 
plt.show()
 
print('The details for confusion matrix is =')
print (classification_report(y_test, y_pred))
```
